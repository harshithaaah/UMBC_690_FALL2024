```python
import pandas as pd
import plotly.express as px

pd.set_option("display.max_rows", None)

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

df.shape
```

```python
df.sample(5)
```

```python
df.isna().sum()
```

```python
df[["Year", "Country","GDP per capita (current US$)", "Life expectancy at birth, total (years)"]].sample(5)
```

```python
df2 = df[df["Year"] != 2022]
df2.shape
```

```python
df2.isna().sum()
```

```python
df3 = df2.dropna()
df3.isna().sum()
```

```python
df3.shape
```

## Homework Question 1:

to check if this dataset has duplicate rows. If it has, only keep one from the duplicated rows and drop the rest.

```python
# Check for duplicates
duplicates = df3.duplicated().sum()

# Print the number of duplicate rows
print(f"Number of duplicate rows: {duplicates}")

# Drop duplicates
df3 = df3.drop_duplicates()
print(f"New shape after dropping duplicates: {df3.shape}")
```

## HW Question 2

Find out if there is a relationship between wealth and health

Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.

```python
# Select data for a specific year, e.g., 2020
data_2020 = df3[df3["Year"] == 2020]

# Create a scatter plot
fig = px.scatter(data_2020,
                 x="GDP per capita (current US$)",
                 y="Life expectancy at birth, total (years)",
                 title="GDP vs Life Expectancy (2020)",
                 labels={
                     "GDP per capita (current US$)": "GDP per capita (US$)",
                     "Life expectancy at birth, total (years)": "Life Expectancy (Years)"
                 })
fig.show()
```

## Homework Question 3

Select a year of data. calculate total population of each region and make a pie chart to display the distribution of total population of all regions.

```python


# Group by Region and sum the population using the correct column name
population_by_region = data_2020.groupby("Region")["Population, total"].sum().reset_index()

# Create a pie chart to display population distribution by region
fig = px.pie(population_by_region,
             values="Population, total",
             names="Region",
             title="Total Population Distribution by Region (2020)",
             hole=0.3)  # Optional: creates a donut chart

# Show the pie chart
fig.show()
```

## Exercise

Calculate average GDP per capita for  each region and make a bar chart.

```python
# Calculate average GDP per capita by region
avg_gdp_by_region = df3.groupby("Region")["GDP per capita (current US$)"].mean().reset_index()

# Create a bar chart
fig = px.bar(avg_gdp_by_region,
              x="Region",
              y="GDP per capita (current US$)",
              title="Average GDP per Capita by Region",
              labels={
                  "GDP per capita (current US$)": "Average GDP per Capita (US$)"
              })
fig.show()
```

## Home work question 4

1. remove the color legend.
2. Order the above bar chart based on the values of the average GDP per capita.

```python
df_group = df_group.sort_values(by="GDP per capita (current US$)", ascending=False)
fig = px.bar(df_group,
             x="Region",
             y="GDP per capita (current US$)",
             title="Average GDP per Capita by Region (2020)")

fig.update_traces(showlegend=False)
fig.show()
```
