## 10/9/20204 Homework Question 1:

for the Quad time series comparison chart above, try to add country name to the end of the line and then remove the color legend.

```python

# Filter data for the selected countries
df_quad = df[df["Country Code"].isin(["USA", "CHN", "IND", "RUS"])]

# Create a line chart
fig = px.line(
    df_quad,
    x="Year",
    y="GDP per capita (current US$)",
    color="Country",
    title="GDP per Capita Over Time for USA, China, India, and Russia",
    labels={
        "GDP per capita (current US$)": "GDP per Capita (Current US$)"
    }
)

# Remove the legend
fig.update_layout(showlegend=False)

# Add country names at the end of each line
for country in df_quad["Country"].unique():
    country_data = df_quad[df_quad["Country"] == country]
    last_year = country_data["Year"].max()
    last_value = country_data[country_data["Year"] == last_year]["GDP per capita (current US$)"].values[0]
    fig.add_annotation(
        x=last_year,
        y=last_value,
        text=country,
        showarrow=False,
        xshift=5  # Slightly shift the text for better positioning
    )

# Show the figure
fig.show()
```

## Question 3 : Bonus

```python
# Group by Year and Region, calculate average GDP per capita, average life expectancy, and total population
df_region_agg = df.groupby(["Year", "Region"]).agg(
    {
        "GDP per capita (current US$)": "mean",
        "Life expectancy at birth, total (years)": "mean",
        "Population, total": "sum"
    }
).reset_index()

df_region_agg.sample(5)
```

```python
# Create a line chart showing the trend of Life Expectancy over time for all regions
fig = px.line(
    df_region_agg,
    x="Year",
    y="Life expectancy at birth, total (years)",
    color="Region",
    title="Life Expectancy Trend for All Regions",
    labels={
        "Year": "Year",
        "Life expectancy at birth, total (years)": "Life Expectancy (Years)"
    }
)

fig.show()


```

```python
# Pick a year from the data and create a scatter plot to compare their GDP per capita and life expectancy.
year_of_interest = 2019
df_region_agg_year = df_region_agg[df_region_agg["Year"] == year_of_interest]

fig = px.scatter(
    df_region_agg_year,
    x="GDP per capita (current US$)",
    y="Life expectancy at birth, total (years)",
    color="Region",
    size="Population, total",
    title=f"GDP per Capita vs. Life Expectancy for All Regions in {year_of_interest}",
    labels={
        "GDP per capita (current US$)": "GDP per Capita (Current US$)",
        "Life expectancy at birth, total (years)": "Life Expectancy (Years)"
    }
)

fig.show()
```

## 10/09/2024 Homework Question 2

Round the Average GDP for each region to 2 decimal points.
Sort the bar according to GDP either in ascending order. Currently, it is ordered alphabetically.

```python

# Group by region to calculate the mean GDP per capita and round it to 2 decimal places
gdp_by_region = df_filtered.groupby('Region').agg(
    {
        'GDP per capita (current US$)': 'mean'
    }
).reset_index()

# Round the 'GDP per capita (current US$)' to 2 decimal places
gdp_by_region['GDP per capita (current US$)'] = gdp_by_region['GDP per capita (current US$)'].round(2)

# Sort the DataFrame by 'GDP per capita (current US$)' in ascending order
gdp_by_region = gdp_by_region.sort_values(by='GDP per capita (current US$)', ascending=True)

# Create a bar chart
fig = px.bar(
    data_frame=gdp_by_region,
    x='Region',
    y='GDP per capita (current US$)',
    color='Region',
    title='Average GDP per Capita by Region',
    labels={
        'GDP per capita (current US$)': 'Average GDP per Capita (Current US$)',
        'Region': 'Region'
    },
    text='GDP per capita (current US$)'
)

# Show the figure
fig.show()
```

## 10/09/2024 Homework question 3

Group the dataset by Region and Income Group, calculate the average, min, and max of GDP per Capita & life Expentancy, total, average, min, and max of population, then save the result to Excel.


```python


import pandas as pd

# Assuming df is already defined with the relevant columns and data
# Define the functions for 25th and 75th percentiles
def q1(x):
    return x.quantile(0.25)

def q3(x):
    return x.quantile(0.75)

# Group the dataset by 'Region' and 'Income Group' and calculate the required statistics
df_grouped = df_2018.groupby(["Region", "Income Group"]).aggregate(
    {
        'GDP per capita (current US$)': ['min', 'mean', 'max'],
        'Life expectancy at birth, total (years)': ['min', 'mean', 'max'],
        'Population, total': ['sum', 'mean', 'min', 'max']
    }
)

# Rename columns for better readability
df_grouped.columns = [
    'GDP per Capita - Minimum', 'GDP per Capita - Average', 'GDP per Capita - Maximum',
    'Life Expectancy - Minimum', 'Life Expectancy - Average', 'Life Expectancy - Maximum',
    'Population - Total', 'Population - Average', 'Population - Minimum', 'Population - Maximum'
]

# Save the result to an Excel file
df_grouped.to_excel("by_region_and_income_group.xlsx")

# Display the first few rows of the result
df_grouped.head()
```
